```python
import pandas as pd
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
import matplotlib.pyplot as plt
```

```python
df = fetch_california_housing(as_frame=True).frame
```

```python
df.describe() # this gets all the statistical information about the data frame
```

```python
df.HouseAge # this will get a Series of a specific columns, in this case the House Age column
```

```python
df.HouseAge.mean()
```

```python
df.HouseAge.std()
```

```python
df['HouseAge'].hist() # this will create a histogram of specific columns
plt.title('House Age')
```

```python
df.hist(figsize=(12,8)) # you can also create an histogram of the full data frame
plt.tight_layout()
```

```python
df.plot(figsize=(12,8))
plt.tight_layout()
```
